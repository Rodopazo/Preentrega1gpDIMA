```python
# Importar la biblioteca OpenAI
import openai

# Configurar la clave API
openai.api_key = "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"  # Reemplaza con tu clave API v√°lida

# Mostrar mensaje de bienvenida
print("üå† Bienvenido al Diccionario M√°gico para Atrevidos üå†")
print("‚ú® Aqu√≠ puedes invocar palabras para descubrir su esencia oculta, su energ√≠a y alternativas m√°gicas. ‚ú®")

# Pedir al usuario que ingrese una palabra
palabra = input("üîÆ ¬øQu√© palabra deseas invocar para conocer su esencia oculta? Escribe tu palabra aqu√≠: ").strip().lower()

# Crear el prompt m√°gico para el an√°lisis textual
concept_prompt = f"""
Analiza la palabra '{palabra}' con base en los siguientes criterios:
1. **Etimolog√≠a**: Describe el origen hist√≥rico de la palabra de manera breve pero completa.
2. **Energ√≠a**: Clasifica su energ√≠a como baja, media o alta, seg√∫n estos criterios:
   - Baja vibraci√≥n: Asociada con miedo, inseguridad, sufrimiento o connotaciones negativas.
   - Media vibraci√≥n: Conceptos neutros o pr√°cticos, sin emociones extremas.
   - Alta vibraci√≥n: Asociada con emociones positivas, entusiasmo o prop√≥sito elevado.
   Proporciona una breve explicaci√≥n de por qu√© tiene esa clasificaci√≥n.
3. **Alternativas m√°gicas**: Sugiere tres palabras alternativas con connotaciones m√°s positivas o vibraciones altas. Explica brevemente cada una.
"""

try:
    # Solicitar la respuesta textual al modelo GPT-4
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "Eres un asistente m√°gico que eval√∫a palabras de manera estructurada: etimolog√≠a, energ√≠a y alternativas, seg√∫n criterios de vibraci√≥n energ√©tica."},
            {"role": "user", "content": concept_prompt}
        ],
        max_tokens=300,  # Ajustado para permitir respuestas completas
        temperature=0.7
    )
    
    # Procesar la respuesta textual
    respuesta = response['choices'][0]['message']['content']
    
    # Dividir y validar cada secci√≥n
    try:
        etimologia = respuesta.split("2. **Energ√≠a**:")[0].strip()
        energia = respuesta.split("2. **Energ√≠a**:")[1].split("3. **Alternativas m√°gicas**:")[0].strip()
        alternativas = respuesta.split("3. **Alternativas m√°gicas**:")[1].strip()
    except IndexError:
        etimologia = "No se encontr√≥ informaci√≥n sobre la etimolog√≠a."
        energia = "No se encontr√≥ informaci√≥n sobre la energ√≠a."
        alternativas = "No se encontraron alternativas m√°gicas."

    # Mostrar el an√°lisis textual
    print("\nüîÆ Aqu√≠ tienes el significado m√°gico de tu palabra:\n")
    print("1. **Etimolog√≠a**:")
    print(etimologia)
    print("\n2. **Energ√≠a**:")
    print(energia)
    print("\n3. **Alternativas m√°gicas y sus significados**:")
    print(alternativas)
    
    # Crear un prompt para la generaci√≥n de im√°genes basado en la etimolog√≠a y energ√≠a
    image_prompt = f"""
    A magical and artistic representation of the word '{palabra}', inspired by:
    - Etimolog√≠a: {etimologia.strip()}
    - Energ√≠a: {energia.strip()}
    Use vibrant colors, fantasy elements, and mystical designs to visually represent its essence.
    """
    
    # Limitar la longitud del prompt de la imagen a 1000 caracteres
    if len(image_prompt) > 1000:
        image_prompt = image_prompt[:1000]

    # Solicitar una imagen al modelo DALL¬∑E
    try:
        image_response = openai.Image.create(
            prompt=image_prompt,
            n=1,
            size="512x512"
        )
        # Obtener la URL de la imagen generada
        image_url = image_response['data'][0]['url']
        print("\nüñºÔ∏è Aqu√≠ tienes una representaci√≥n visual m√°gica de tu palabra:\n")
        print(image_url)
    except Exception as img_error:
        print("\n‚ö†Ô∏è No se pudo generar la imagen. Detalles del error:")
        print(str(img_error))

except Exception as e:
    print("\n‚ö†Ô∏è Hubo un error al procesar tu solicitud. Por favor, verifica tu conexi√≥n o clave API.")
    print(str(e))
```
